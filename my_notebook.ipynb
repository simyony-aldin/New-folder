```python
# תחזית תוצאות עסקיות
print("="*80)
print("💰 תחזית תוצאות עסקיות - ROI Analysis")
print("="*80)

# נתונים לדוגמה (יש להתאים למציאות הארגונית)
current_metrics = {
    "סה\"כ לקוחות": 8950,
    "Retention Rate (%)": 85,
    "Avg Revenue per User ($)": 100,
    "מחיקות חובות שנתיות ($M)": 2.0,
    "Customer Satisfaction": 7.5,
    "ROI על שיווק": "3:1"
}

target_metrics = {
    "סה\"כ לקוחות": 9400,  # +5% growth
    "Retention Rate (%)": 90,
    "Avg Revenue per User ($)": 120,
    "מחיקות חובות שנתיות ($M)": 1.5,
    "Customer Satisfaction": 8.5,
    "ROI על שיווק": "5:1"
}

print("\n📊 השוואת מדדים - נוכחי vs יעד (שנה ראשונה):")
print("-" * 80)
print(f"{'מדד':<40} {'נוכחי':>15} {'יעד':>15} {'שיפור':>10}")
print("-" * 80)

improvements = []

for metric in current_metrics.keys():
    current = current_metrics[metric]
    target = target_metrics[metric]
    
    if isinstance(current, (int, float)) and isinstance(target, (int, float)):
        if current != 0:
            improvement = ((target - current) / current) * 100
            improvements.append({
                'metric': metric,
                'current': current,
                'target': target,
                'improvement': improvement
            })
            print(f"{metric:<40} {current:>15.1f} {target:>15.1f} {improvement:>9.1f}%")
        else:
            print(f"{metric:<40} {current:>15} {target:>15} {'N/A':>10}")
    else:
        print(f"{metric:<40} {str(current):>15} {str(target):>15} {'':>10}")

print("-" * 80)

# חישוב השפעה פיננסית
print("\n💵 השפעה פיננסית משוערת (שנה ראשונה):")
print("-" * 80)

revenue_increase = (target_metrics["Avg Revenue per User ($)"] - current_metrics["Avg Revenue per User ($)"]) * current_metrics["סה\"כ לקוחות"]
retention_value = (target_metrics["Retention Rate (%)"] - current_metrics["Retention Rate (%)"]) / 100 * current_metrics["סה\"כ לקוחות"] * current_metrics["Avg Revenue per User ($)"]
risk_reduction = (current_metrics["מחיקות חובות שנתיות ($M)"] - target_metrics["מחיקות חובות שנתיות ($M)"]) * 1000000

total_impact = revenue_increase + retention_value + risk_reduction

print(f"1. הגדלת Revenue per User:        ${revenue_increase:>15,.0f}")
print(f"2. שיפור Retention:                ${retention_value:>15,.0f}")
print(f"3. הפחתת מחיקות חובות:           ${risk_reduction:>15,.0f}")
print("-" * 80)
print(f"סה\"כ השפעה פיננסית:              ${total_impact:>15,.0f}")
print("=" * 80)

print(f"\n🎯 תוצאה צפויה: ${total_impact/1000000:.1f}M רווח נוסף בשנה הראשונה")
print(f"💡 ROI: {total_impact/75000:.1f}x (בהנחת השקעה של $75K)")

# פירוט לפי קבוצה
print("\n" + "="*80)
print("📈 פירוט השפעה לפי קבוצת לקוחות:")
print("="*80)

if 'Cluster' in df.columns:
    for cluster_id in sorted(df['Cluster'].unique()):
        cluster_data = df[df['Cluster'] == cluster_id]
        cluster_size = len(cluster_data)
        cluster_pct = cluster_size / len(df) * 100
        
        # הערכת פוטנציאל לפי מאפייני הקבוצה
        avg_purchases = cluster_data['PURCHASES'].mean()
        
        if avg_purchases > 2000:
            potential = "⭐⭐⭐⭐⭐ (VIP - שימור קריטי)"
            impact = f"${cluster_size * 1000:,.0f}/שנה"
        elif avg_purchases > 1000:
            potential = "⭐⭐⭐⭐ (גבוה)"
            impact = f"${cluster_size * 200:,.0f}/שנה"
        else:
            potential = "⭐⭐⭐ (בינוני)"
            impact = f"${cluster_size * 100:,.0f}/שנה"
        
        print(f"\nCluster {cluster_id} ({cluster_pct:.1f}% מהלקוחות):")
        print(f"   פוטנציאל: {potential}")
        print(f"   השפעה משוערת: {impact}")
else:
    print("⚠️ לא נמצאה עמודת Cluster")
```

### ג. תוכנית יישום - Roadmap

**לוח זמנים ליישום המודל בארגון:**

```python
# מערכת ניהול סיכונים מבוססת Clustering
print("="*80)
print("🛡️ ניהול סיכוני אשראי - Risk Management Framework")
print("="*80)

if 'Cluster' in df.columns:
    # חישוב רמות סיכון לכל cluster
    risk_analysis = []
    
    for cluster_id in sorted(df['Cluster'].unique()):
        cluster_data = df[df['Cluster'] == cluster_id]
        
        # מדדי סיכון
        avg_balance = cluster_data['BALANCE'].mean()
        avg_cash_advance = cluster_data['CASH_ADVANCE'].mean()
        avg_payment_rate = cluster_data['PRC_FULL_PAYMENT'].mean()
        avg_credit_limit = cluster_data['CREDIT_LIMIT'].mean()
        
        # חישוב ציון סיכון (0-100)
        # גבוה יותר = סיכון גבוה יותר
        risk_score = 0
        
        # משיכות מזומן גבוהות = סיכון
        if avg_cash_advance > 2000:
            risk_score += 40
        elif avg_cash_advance > 1000:
            risk_score += 20
            
        # תשלום מלא נמוך = סיכון
        if avg_payment_rate < 0.1:
            risk_score += 30
        elif avg_payment_rate < 0.3:
            risk_score += 15
            
        # יתרה גבוהה יחסית למסגרת = סיכון
        utilization_rate = avg_balance / avg_credit_limit if avg_credit_limit > 0 else 0
        if utilization_rate > 0.8:
            risk_score += 30
        elif utilization_rate > 0.6:
            risk_score += 15
        
        # קטגוריית סיכון
        if risk_score >= 60:
            risk_category = "🔴 סיכון גבוה"
            action = "ניטור יומי + התערבות מיידית"
        elif risk_score >= 30:
            risk_category = "🟡 סיכון בינוני"
            action = "ניטור שבועי + התראות"
        else:
            risk_category = "🟢 סיכון נמוך"
            action = "ניטור רגיל"
        
        risk_analysis.append({
            'Cluster': cluster_id,
            'גודל': len(cluster_data),
            'אחוז': f"{len(cluster_data)/len(df)*100:.1f}%",
            'ציון_סיכון': risk_score,
            'קטגוריה': risk_category,
            'פעולה': action,
            'משיכות_מזומן_ממוצעות': f"${avg_cash_advance:,.0f}",
            'תשלום_מלא': f"{avg_payment_rate*100:.1f}%",
            'ניצול_מסגרת': f"{utilization_rate*100:.1f}%"
        })
    
    # הצגה
    risk_df = pd.DataFrame(risk_analysis)
    print("\n📊 מטריצת סיכונים:")
    print(risk_df.to_string(index=False))
    
    # המלצות
    print("\n" + "="*80)
    print("🎯 המלצות לניהול סיכונים:")
    print("="*80)
    
    high_risk = risk_df[risk_df['ציון_סיכון'] >= 60]
    if len(high_risk) > 0:
        print(f"\n🔴 סיכון גבוה: {len(high_risk)} קבוצות")
        for _, row in high_risk.iterrows():
            print(f"   • Cluster {row['Cluster']} ({row['אחוז']}) - {row['פעולה']}")
            print(f"     → חיסכון פוטנציאלי: $500K-$1M בשנה (הפחתת מחיקות)")
    
    medium_risk = risk_df[(risk_df['ציון_סיכון'] >= 30) & (risk_df['ציון_סיכון'] < 60)]
    if len(medium_risk) > 0:
        print(f"\n🟡 סיכון בינוני: {len(medium_risk)} קבוצות")
        for _, row in medium_risk.iterrows():
            print(f"   • Cluster {row['Cluster']} ({row['אחוז']}) - {row['פעולה']}")
    
    low_risk = risk_df[risk_df['ציון_סיכון'] < 30]
    if len(low_risk) > 0:
        print(f"\n🟢 סיכון נמוך: {len(low_risk)} קבוצות")
        print(f"   ✅ ניטור רגיל מספיק - אלו הלקוחות היציבים")

else:
    print("⚠️ לא נמצאה עמודת Cluster")
```

### החלטה על K אופטימלי

**ניתוח הגרף:**

מסתכלים על הגרף ומחפשים את נקודת ה"מרפק" - המקום שבה הירידה מתחילה להאט משמעותית.

בדרך כלל:
- **K=3-4**: מתאים לפילוח בסיסי (נמוך, בינוני, גבוה)
- **K=4-5**: מאפשר הבחנות מפורטות יותר
- **K>6**: עלול להיות overfitting (יותר מדי קבוצות קטנות)

**לצורכי הדוגמה - אני בוחר K=4** כי:
1. יש ירידה משמעותית עד K=4
2. אחרי K=4 הירידה מתונה יותר
3. זה מספיק קבוצות למבצעים שיווקיים משמעותיים
4. לא יותר מדי (קל יותר לנהל)

```python
# ויזואליזציה - Elbow Curve
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertias, marker='o', linestyle='-', linewidth=2, markersize=8, color='blue')
plt.xlabel('מספר Clusters (K)', fontsize=12)
plt.ylabel('Inertia (סכום מרחקים)', fontsize=12)
plt.title('Elbow Method - מציאת K אופטימלי', fontsize=14, fontweight='bold')
plt.xticks(K_range)
plt.grid(True, alpha=0.3)

# סימון נקודות חשובות
for i, k in enumerate(K_range):
    plt.text(k, inertias[i], f'{inertias[i]:,.0f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

print("\n🔍 איך לזהות את ה-Elbow?")
print("   1. מחפשים נקודה שבה הגרף 'נשבר' - כאן הירידה הופכת פחות תלולה")
print("   2. לפני הנקודה: הירידה חדה (הוספת cluster עוזרת הרבה)")
print("   3. אחרי הנקודה: הירידה מתונה (הוספת cluster עוזרת מעט)")
```

### גרף Elbow - זיהוי ה"מרפק"

```python
# Elbow Method - חישוב Inertia עבור K שונים
print("🔍 מחשב Elbow Method...")
print("="*70)

inertias = []
K_range = range(2, 11)  # נבדוק K=2 עד K=10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10, max_iter=300)
    kmeans.fit(df_scaled)
    inertias.append(kmeans.inertia_)
    print(f"   K={k}: Inertia = {kmeans.inertia_:,.2f}")

print("\n✅ חישוב הושלם!")
print("\n💡 הסבר: Inertia = סכום המרחקים של כל נקודה מהמרכז של הקבוצה שלה")
print("   ככל ש-K גדל יותר, Inertia קטן (קבוצות קטנות יותר)")
```

### שלב 2.2: ויזואליזציה - היסטוגרמות

**למה היסטוגרמות?**
- רוצים לראות איך הנתונים מתפלגים
- לזהות א-סימטריה (skewness)
- למצוא ערכים קיצוניים (outliers)

```python
# בחירת עמודות עיקריות לניתוח קורלציה
key_features = ['BALANCE', 'PURCHASES', 'CASH_ADVANCE', 'CREDIT_LIMIT', 'PAYMENTS']

# חישוב מטריצת קורלציה
correlation_matrix = df[key_features].corr()

print("📊 מטריצת קורלציה בין המשתנים העיקריים:")
print("="*70)
print(correlation_matrix.round(3))

# ויזואליזציה - Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('מטריצת קורלציה - משתנים עיקריים', fontsize=16, pad=20)
plt.tight_layout()
plt.show()

print("\n💡 פרשנות:")
print("   🔴 אדום כהה (קרוב ל-1): קשר חזק חיובי")
print("   🔵 כחול כהה (קרוב ל-1-): קשר חזק שלילי")
print("   ⚪ לבן (קרוב ל-0): אין קשר")
```

### שלב 2.1: מטריצת קורלציה (Correlation Matrix)

**מה זה קורלציה?**
- מדד לקשר בין שני משתנים (-1 עד 1)
- קורלציה חיובית (קרוב ל-1): כשאחד עולה, השני עולה
- קורלציה שלילית (קרוב ל-1-): כשאחד עולה, השני יורד
- קורלציה 0: אין קשר

### מטרה:
להבין את הקשרים בין המשתנים, לזהות outliers ולראות את התפלגות הנתונים

```python
# טיפול בערכים חסרים - השלמה בחציון
print("🔧 טיפול בערכים חסרים:")
print("="*70)

# שמירת מספר הערכים החסרים לפני
missing_before = df.isnull().sum().sum()

# השלמת ערכים חסרים בחציון של כל עמודה
for col in df.select_dtypes(include=[np.number]).columns:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"✅ {col}: הושלם ב-{median_val:.2f} (חציון)")

# בדיקה לאחר הטיפול
missing_after = df.isnull().sum().sum()

print(f"\n📊 סיכום:")
print(f"   לפני: {missing_before:,} ערכים חסרים")
print(f"   אחרי: {missing_after:,} ערכים חסרים")
print(f"   ✅ טופלו {missing_before - missing_after:,} ערכים")
```

### שלב 1.5: טיפול בערכים חסרים

**החלטה:** אני משתמש ב-**Median (חציון)** במקום Mean (ממוצע) כי:
1. הנתונים הפיננסיים לרוב לא נורמליים (יש outliers)
2. החציון עמיד יותר לערכים קיצוניים
3. זה שיטה מקובלת בנתונים פיננסיים

```python
# בדיקת ערכים חסרים
print("🔍 בדיקת ערכים חסרים (NaN):")
print("="*70)

missing = pd.DataFrame({
    'עמודה': df.columns,
    'ערכים חסרים': df.isnull().sum(),
    'אחוז חסר (%)': (df.isnull().sum() / len(df) * 100).round(2)
})

missing = missing[missing['ערכים חסרים'] > 0].sort_values('ערכים חסרים', ascending=False)

if len(missing) > 0:
    print(missing.to_string(index=False))
    print(f"\n⚠️ סה״כ {len(missing)} עמודות עם ערכים חסרים")
else:
    print("✅ אין ערכים חסרים בנתונים!")
```
